```python
df = pd.read_csv("flights.csv")

# Group by Year and Month with aggregations
flights_grouped = df.groupby(["Year", "Month"]).agg(
    Total_Flights=("FlightNum", "count"),
    Avg_Arrival_Delay=("ArrDelay", "mean"),
    Max_Departure_Delay=("DepDelay", "max")
).reset_index()

# Display results
print(flights_grouped)
```

```python
df = pd.read_excel("titanic.xlsx")
```

```python
def classify_age(age):
    if pd.isnull(age):  # Handle missing values
        return "Unknown"
    return "Child" if age < 18 else "Adult"

# Apply the function 
df["Age_Group"] = df["Age"].apply(classify_age)

# Display the first few rows
print(df[["Age", "Age_Group"]].head())
```

```python
df.head()
```

```python
# Define a function for Min-Max normalization
def min_max_normalize(series):
    return (series - series.min()) / (series.max() - series.min())
```

```python
df["Normalized_Salary"] = df.groupby("DEPARTMENT")["BASE_SALARY"].transform(min_max_normalize)

# Display the first few rows
print(df[["DEPARTMENT", "BASE_SALARY", "Normalized_Salary"]].head())
```

```python
def classify_duration(minutes):
    if minutes < 60:
        return "Short"
    elif 60 <= minutes <= 120:
        return "Medium"
    else:
        return "Long"
```

```python
df["Movie_Length"] = df["duration"].apply(classify_duration)

# Display the first few rows
print(df[["director_name", "duration", "Movie_Length"]].head())
```

```python
df = pd.read_excel("titanic.xlsx")
```

```python
def preprocess_titanic(df):
    # Filter passengers who survived
    df = df[df["Survived"] == 1].copy()

    # Fill missing Age values with the mean
    df["Age"].fillna(df["Age"].mean(), inplace=True)

    # Create the new column
    df["Fare_Per_Age"] = df["Fare"] / df["Age"]

    return df
```

```python
df_processed = preprocess_titanic(df)

# Display the first few rows
print(df_processed[["Survived", "Age", "Fare", "Fare_Per_Age"]].head())
```

```python
df = pd.read_csv("flights.csv")

# Define the preprocessing function
def preprocess_flights(df):
    # Filter flights with a departure delay > 30 minutes
    df = df[df["DepartureDelay"] > 30].copy()

    # Avoid division by zero and create the new column
    df["Delay_Per_Hour"] = df["DepartureDelay"] / df["ScheduledDuration"]

    return df

# Apply the function
df_processed = preprocess_flights(df)

# Display the first few rows
print(df_processed[["DepartureDelay", "ScheduledDuration", "Delay_Per_Hour"]].head())
```
