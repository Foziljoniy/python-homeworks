```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
# 1. Basic Plotting
x = np.linspace(-10, 10, 400)
y = x**2 - 4*x + 4

plt.figure(figsize=(6, 4))
plt.plot(x, y, label='$f(x) = x^2 - 4x + 4$', color='b')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Basic Plotting')
plt.legend()
plt.grid()
```

```python
# 2. Sine and Cosine Plot
x = np.linspace(0, 2*np.pi, 400)
sin_x = np.sin(x)
cos_x = np.cos(x)

plt.figure(figsize=(6, 4))
plt.plot(x, sin_x, label='$\sin(x)$', linestyle='--', color='r')
plt.plot(x, cos_x, label='$\cos(x)$', linestyle='-.', color='g')
plt.xlabel('x')
plt.ylabel('Function value')
plt.title('Sine and Cosine Plot')
plt.legend()
plt.grid()
plt.show()
```

```python
# 3. Subplots
x = np.linspace(-2, 2, 400)
x_positive = np.linspace(0, 2, 400)  # For log function

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Top-left: f(x) = x^3
axs[0, 0].plot(x, x**3, 'b')
axs[0, 0].set_title('$f(x) = x^3$')
axs[0, 0].grid()

# Top-right: f(x) = sin(x)
axs[0, 1].plot(x, np.sin(x), 'r')
axs[0, 1].set_title('$f(x) = \sin(x)$')
axs[0, 1].grid()

# Bottom-left: f(x) = e^x
axs[1, 0].plot(x, np.exp(x), 'g')
axs[1, 0].set_title('$f(x) = e^x$')
axs[1, 0].grid()

# Bottom-right: f(x) = log(x+1)
axs[1, 1].plot(x_positive, np.log(x_positive + 1), 'm')
axs[1, 1].set_title('$f(x) = \log(x+1)$ (for $x \geq 0$)')
axs[1, 1].grid()

plt.tight_layout()
plt.show()
```

```python
# 4. Scatter Plot
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)
colors = np.random.rand(100)
plt.scatter(x, y, c=colors, cmap='viridis', marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.colorbar(label='Color Intensity')
plt.grid()
plt.show()
```

```python
# 5. Histogram
data = np.random.normal(0, 1, 1000)
plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.grid()
plt.show()
```

```python
# 6. 3D Plotting
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.cos(X**2 + Y**2)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm')
fig.colorbar(surf)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')
ax.set_title('3D Surface Plot')
plt.show()
```

```python
# 7. Bar Chart
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [200, 150, 250, 175, 225]
plt.bar(products, sales, color=['blue', 'green', 'red', 'purple', 'orange'])
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Sales Data')
plt.show()
```

```python
# 8. Stacked Bar Chart
categories = ['Category A', 'Category B', 'Category C']
time_periods = ['T1', 'T2', 'T3', 'T4']
data = np.random.randint(10, 100, (3, 4))
fig, ax = plt.subplots()
ax.bar(time_periods, data[0], label='Category A')
ax.bar(time_periods, data[1], bottom=data[0], label='Category B')
ax.bar(time_periods, data[2], bottom=data[0]+data[1], label='Category C')
ax.set_xlabel('Time Periods')
ax.set_ylabel('Value')
ax.set_title('Stacked Bar Chart')
ax.legend()
plt.show()
```
